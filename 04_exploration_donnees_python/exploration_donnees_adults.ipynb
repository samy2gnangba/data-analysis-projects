```python
# 1. CONNEXION AU DATASET
```

```python
# Importer la bibliothèque pandas
# On utilise 'as pd' pour pouvoir écrire 'pd'
import pandas as pd
```

```python
# 1. On crée une liste avec le nom des colonnes
noms_colonnes = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
                 'marital_status', 'occupation', 'relationship', 'race', 'sex',
                 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']

# 2. On charge le fichier en utilisant le chemin d'accès
chemin = "/content/adults.csv"
df = pd.read_csv(chemin, names=noms_colonnes, skipinitialspace=True)
```

```python
# Afficher les 5 premières lignes
print(df.head())
```

```python
# 2. INFORMATIONS GÉNÉRALES
```

```python
# Affiche le nombre de lignes et de colonnes
print(df.shape)
```

```python
# Affiche le type de chaque colonne
print(df.dtypes)
```

```python
# Affiche le nombre de valeurs manquantes par colonne
print(df.isnull().sum())
```

```python
# 3. Statistiques descriptives globales
```

```python
# On génère le tableau des statistiques descriptives
statistiques = df.describe()

# On affiche le résultat
print(statistiques)
```

```python
# 4. EXPLORATION D'UNE VARIABLE (EDUCATION)
```

```python
# On compte le nombre d'apparitions de chaque catégorie dans la colonne 'education'
distribution_education = df['education'].value_counts()

# On affiche le résultat
print("Distribution du niveau d'education :")
print(distribution_education)
```

```python
# 5. FILTRAGE SIMPLE
```

```python
# On crée un nouveau tableau appelé 'resultat'
# On demande comme condition: âge > 40 ET revenu == '>50K'
resultat = df[(df['age'] > 40) & (df['income'] == '>50K')]

# On affiche les premières lignes
print("Voici les personnes de plus de 40 ans gagnant plus de 50K :")
print(resultat.head())

# On affiche le nombre total de personnes trouvées
print("\nNombre de personnes trouvées :")
print(len(resultat))
```
